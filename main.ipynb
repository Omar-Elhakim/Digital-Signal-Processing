```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
fileName = "SinOutput.txt"

with open(fileName, "r") as f:
    timeFlag = f.readline()
    periodicFlag = f.readline()

    nOfSamples = int(f.readline())
    indices = []
    amplitudes = []
    for _ in range(nOfSamples):
        line = f.readline().strip().split(" ")
        indices.append(int(line[0]))
        amplitudes.append(float(line[1]))
```

```python
fig, ax = plt.subplots(figsize=(10, 5))
ax.plot(indices, amplitudes)

# st.pyplot(fig) # for streamlit
plt.show(fig)  # for notebook
```
